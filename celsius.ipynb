```python
import pandas as pd
df = pd.read_csv('1A_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
#plt.plot(df['Timestamp'],df['Temperature_Celsius(°C)'])
#plt.plot(df['Timestamp'],df['Relative_Humidity(%)'])
from matplotlib import cm
#import japanize_matplotlib
%matplotlib inline
ffig, ax1 = plt.subplots()
ax2 = ax1.twinx()
# 2軸グラフの本体設定
ax1.plot(df['Timestamp'],df['Temperature_Celsius(°C)'],
        color=cm.Set1.colors[1], label="温度")
ax2.plot(df['Timestamp'],df['Relative_Humidity(%)'],
        color=cm.Set1.colors[0], alpha=0.4, label="相対湿度")
plt.tick_params(labelsize = 10) #目盛線ラベルのフォントサイズ
#グラフタイトルを付ける
plt.title("温度と湿度の推移", fontsize=15)
# 凡例の表示のため、handler1と2にはグラフオブジェクトのリスト情報が入る
# label1と2には、凡例用に各labelのリスト情報が入る
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
# 凡例をまとめて出力する
ax1.legend(handler1 + handler2, label1 + label2, loc=2, borderaxespad=0.)
temperature_max = 10 + max(df["Temperature_Celsius(°C)"])
rainfall_max = 1.2 * max(df["Relative_Humidity(%)"])
ax1.set_ylim([0, temperature_max])
ax2.set_ylim([0, rainfall_max])
plt.legend(loc = 'upper right')
plt.show()
```
